```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# --- 1. DYNAMIC PATH CONFIGURATION ---
current_location = Path.cwd()

# Logic to find project root
if current_location.name == "notebooks":
    PROJECT_ROOT = current_location.parent
elif (current_location / "data").exists():
    PROJECT_ROOT = current_location
else:
    PROJECT_ROOT = current_location.parent # Fallback

DATA_DIR = PROJECT_ROOT / "data"

PATHS = {
    "Real_Dataset": DATA_DIR / "raw_real",       
    "Synthetic_Dataset": DATA_DIR / "raw_sintetico"   
}

# --- 2. SCANNING FUNCTION ---
def scan_dataset(dataset_name, root_path):
    data = []
    image_extensions = {'.jpg', '.jpeg', '.png', '.bmp'}
    all_images = []
    
    if not root_path.exists():
        print(f"❌ Path not found: {root_path}")
        return pd.DataFrame()

    # Recursive search for images
    for ext in image_extensions:
        all_images.extend(list(root_path.rglob(f"*{ext}")))
        all_images.extend(list(root_path.rglob(f"*{ext.upper()}")))
    
    all_images = sorted(list(set(all_images))) 

    for img_path in all_images:
        # logic to find label file
        label_path = None
        possibles = [
            img_path.with_suffix('.txt'),
            img_path.parents[1] / 'labels' / f"{img_path.stem}.txt", # standard yolo
            img_path.parent / 'labels' / f"{img_path.stem}.txt"      # flat structure
        ]
        
        for p in possibles:
            if p.exists():
                label_path = p
                break
        
        # Determine status: "With Label" vs "Without Label"
        status = "With Label"
        
        if label_path is None:
            status = "Without Label" # Missing txt
        elif label_path.stat().st_size == 0:
            status = "Without Label" # Empty txt
        else:
            try:
                with open(label_path, 'r') as f:
                    if not f.read().strip(): 
                        status = "Without Label" # Text file with whitespace only
            except: 
                pass
        
        data.append({
            'Dataset': dataset_name,
            'Filename': img_path.name,
            'Status': status
        })
        
    return pd.DataFrame(data)

# --- 3. EXECUTION AND VISUALIZATION ---
dfs = []
for name, path in PATHS.items():
    print(f"Scanning {name}...")
    dfs.append(scan_dataset(name, path))

if dfs:
    df_full = pd.concat(dfs)
    
    # --- TABLE 1: SUMMARY BY DATASET ---
    print("\n--- DATASET SUMMARY TABLE ---")
    summary_table = df_full.groupby(['Dataset', 'Status']).size().unstack(fill_value=0)
    summary_table['Total'] = summary_table.sum(axis=1)
    
    # Calculate percentages
    if 'Without Label' in summary_table.columns:
        summary_table['% Without Label'] = (summary_table['Without Label'] / summary_table['Total'] * 100).round(1)
    
    display(summary_table) # Formatted Pandas Table

    # --- PLOT: DISTRIBUTION ---
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df_full, x='Dataset', hue='Status', palette='viridis')
    
    plt.title("Image Distribution: With Label vs Without Label")
    plt.ylabel("Number of Images")
    plt.xlabel("Dataset Source")
    plt.legend(title="Image Type")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Add counts on top of bars
    ax = plt.gca()
    for container in ax.containers:
        ax.bar_label(container)
        
    plt.show()

else:
    print("❌ No images found in the specified paths.")
```

```python
import sys
from pathlib import Path
from tqdm import tqdm
from collections import Counter

# ==============================================================================
# CONFIGURAÇÃO DE CAMINHOS
# ==============================================================================
current_location = Path.cwd()

if current_location.name == "notebooks":
    PROJECT_ROOT = current_location.parent
elif (current_location / "data").exists():
    PROJECT_ROOT = current_location
elif (current_location / "src").exists():
    PROJECT_ROOT = current_location
else:
    PROJECT_ROOT = current_location.parent

DATA_DIR = PROJECT_ROOT / "data"
RAW_REAL_DIR = DATA_DIR / "raw_real"

# Mapeamento de classes (Texto -> Número)
CLASS_MAPPING = {'damage': 0, 'dirt': 1, '0': 0, '1': 1}

def analyze_filtered_dataset():
    print(f"🚀 Iniciando Pente Fino em: {RAW_REAL_DIR}")
    
    if not RAW_REAL_DIR.exists():
        print(f"❌ ERRO: A pasta {RAW_REAL_DIR} não existe!")
        return

    # Buscar todos os arquivos de label
    lbl_files = [f for f in RAW_REAL_DIR.rglob("*.txt") if f.name != "classes.txt"]

    # Contadores
    stats = {
        "total_files": len(lbl_files),
        "empty_images": 0,           # Imagens sem nenhuma anotação (Background puro)
        "annotated_images": 0,       # Imagens com pelo menos 1 box
        "images_with_damage": 0,     # Quantas imagens têm pelo menos 1 dano
        "images_with_dirt": 0,       # Quantas imagens têm sujeira
        "total_damage_boxes": 0,
        "total_dirt_boxes": 0
    }

    print("\n🕵️‍♂️ Analisando conteúdo dos labels...")
    
    for lbl_path in tqdm(lbl_files, desc="Lendo Arquivos"):
        try:
            with open(lbl_path, 'r') as f:
                lines = f.readlines()
                
                # Se o arquivo está vazio, é Background
                if len(lines) == 0:
                    stats["empty_images"] += 1
                    continue

                # Se tem linhas, analisamos o que tem dentro
                has_damage = False
                has_dirt = False
                valid_box_count = 0

                for line in lines:
                    parts = line.strip().split()
                    if len(parts) > 0:
                        raw_cls = parts[0]
                        
                        # Converte para ID numérico
                        if raw_cls.isdigit():
                            cls_id = int(raw_cls)
                        else:
                            cls_id = CLASS_MAPPING.get(raw_cls.lower())
                        
                        if cls_id == 0: # Damage
                            stats["total_damage_boxes"] += 1
                            has_damage = True
                            valid_box_count += 1
                        elif cls_id == 1: # Dirt
                            stats["total_dirt_boxes"] += 1
                            has_dirt = True
                            valid_box_count += 1
                
                if valid_box_count > 0:
                    stats["annotated_images"] += 1
                    if has_damage: stats["images_with_damage"] += 1
                    if has_dirt: stats["images_with_dirt"] += 1

        except Exception as e:
            print(f"Erro ao ler {lbl_path.name}: {e}")

    # ==============================================================================
    # 🖨️ RELATÓRIO CORRIGIDO
    # ==============================================================================
    total_boxes = stats["total_damage_boxes"] + stats["total_dirt_boxes"]
    
    print("\n" + "="*50)
    print("📜 DADOS PRECISOS PARA O TCC")
    print("="*50)
    print(f"1. QUANTITATIVO DE IMAGENS:")
    print(f"   - Total de Arquivos (Labels): {stats['total_files']}")
    print(f"   - Imagens VAZIAS (Background): {stats['empty_images']}")
    print(f"   - Imagens ÚTEIS (Com anotação): {stats['annotated_images']} (Esse é o número chave!)")
    
    print(f"\n2. DISTRIBUIÇÃO DAS IMAGENS ÚTEIS:")
    print(f"   - Imagens contendo DANO: {stats['images_with_damage']}")
    print(f"   - Imagens contendo SUJEIRA: {stats['images_with_dirt']}")
    print(f"   (Nota: Uma imagem pode ter os dois)")

    print(f"\n3. QUANTITATIVO DE OBJETOS (BOXES):")
    print(f"   - Total de Boxes: {total_boxes}")
    print(f"   - Boxes de DANO: {stats['total_damage_boxes']} ({stats['total_damage_boxes']/total_boxes*100:.2f}%)")
    print(f"   - Boxes de SUJEIRA: {stats['total_dirt_boxes']} ({stats['total_dirt_boxes']/total_boxes*100:.2f}%)")
    
    if stats['annotated_images'] > 0:
        avg_objs = total_boxes / stats['annotated_images']
        print(f"\n4. DENSIDADE REAL:")
        print(f"   - Média de objetos por imagem anotada: {avg_objs:.2f}")

if __name__ == "__main__":
    analyze_filtered_dataset()
```
